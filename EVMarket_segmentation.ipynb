```python
from sklearn.preprocessing import MinMaxScaler

# Define a function to scale columns
def scale_column(data, column_name):
    x = data[column_name].values.reshape(-1, 1)
    scaler = MinMaxScaler()
    scaled_values = scaler.fit_transform(x)
    data[column_name] = scaled_values

# List of columns to scale
columns_to_scale = ['Wife Salary', 'Total Salary', 'Price']

# Apply scaling function to each column
for col in columns_to_scale:
    scale_column(d, col)

# Optionally, print the head of the DataFrame after scaling
print(d.head())
```
